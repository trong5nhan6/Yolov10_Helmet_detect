# 7. Show kết quả dự đoán

```python
import matplotlib.pyplot as plt
import cv2

# Đường dẫn đến hai ảnh (ví dụ)
IMG_PATH1 = './images/HCMC_Street.jpg'
IMG_PATH2 = './images/HCMC_Street_predict.png'

# Đọc ảnh bằng OpenCV
img_origin = cv2.imread(IMG_PATH1)
img_predict = cv2.imread(IMG_PATH2)

# Chuyển đổi từ BGR sang RGB (nếu cần)
img_origin = cv2.cvtColor(img_origin, cv2.COLOR_BGR2RGB)
img_predict = cv2.cvtColor(img_predict, cv2.COLOR_BGR2RGB)

# Tạo subplot với 1 hàng và 2 cột
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

# Hiển thị ảnh đầu tiên
axes[0].imshow(img_origin)
axes[0].axis('off')  # Tắt trục đồng thời

# Hiển thị ảnh thứ hai
axes[1].imshow(img_predict)
axes[1].axis('off')  # Tắt trục đồng thời

# Hiển thị lưới ảnh
plt.tight_layout()
plt.show()
```
